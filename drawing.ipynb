# Drawing with Open Source Computer Vision Library

### Cloud

```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt  

img = np.zeros((600,900,3),dtype=np.uint8) 
cv.rectangle(img, (0,0),(900,600),(255,225,85),-1) 
cv.circle(img, (300,350),100,(255,255,255),-1) 
cv.circle(img, (450,350),100,(255,255,255),-1) 
cv.circle(img, (600,350),100,(255,255,255),-1) 
cv.circle(img, (400,250),100,(255,255,255),-1) 
cv.circle(img, (500,250),100,(255,255,255),-1) 

plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
plt.axis('off')  
plt.show()
```

### Person

```python
img = np.zeros((600,900,3),dtype=np.uint8) 

cv.rectangle(img, (0,0),(900,600),(0,47,96),-1) 
cv.circle(img, (450,100),80,(191,191,252),-1) 
cv.line(img, (450,150),(450,400),(191,191,252),5) 
cv.line(img, (450,400),(525,580),(191,191,252),5) 
cv.line(img, (450,400),(375,580),(191,191,252),5) 
cv.line(img, (450,250),(250,150),(191,191,252),5) 
cv.line(img, (450,250),(650,300),(191,191,252),5) 
cv.circle(img, (420,80),8,(0,0,0),-1) 
cv.circle(img, (480,80),8,(0,0,0),-1) 
cv.line(img, (440,120),(460,120),(0,0,191),5) 

plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
plt.axis('off')  
plt.show()
```

### House

```python
img = np.zeros((600,900,3),dtype=np.uint8) 

cv.rectangle(img, (0,0),(900,550),(255,225,85),-1) 
cv.rectangle(img, (0,550),(900,600),(63,127,0),-1) 
cv.rectangle(img, (200,250),(700,550),(212,170,255),-1) 
triangle=np.array([[200,250],[450,20],[700,250]], dtype=np.int32) 
cv.fillPoly(img, [triangle], (0,63,127)) 
cv.rectangle(img, (300,400),(400,550),(51,51,51),-1) 
cv.rectangle(img, (550,300),(650,400),(51,51,51),-1) 
cv.line(img, (600,300),(600,400),(255,255,255),3) 
cv.line(img, (550,350),(650,350),(255,255,255),3) 
cv.circle(img, (380,475),5,(255,255,255),-1)

plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))  # Convert BGR to RGB for matplotlib
plt.axis('off')  
plt.show()
```
